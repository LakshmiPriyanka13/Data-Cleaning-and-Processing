```python
import pandas as pd
```

```python
data = pd.read_csv('movies.csv')
data.head()
```

```python
data.info()
```

```python
null_counts = data.isnull().sum()
print(null_counts)
```

```python
data[["GENRE" ,"ONE-LINE" ,"STARS"]] = data[["GENRE" ,"ONE-LINE" ,"STARS"]].replace('[^a-zA-Z0-9 ]','')
data.head()
```

```python
data["GENRE"]=data["GENRE"].str.replace("\n","").str.strip()
data["ONE-LINE"]=data["ONE-LINE"].str.replace("\n","").str.strip()
data["STARS"]=data["STARS"].str.replace("\n","").str.strip()
data[["GENRE" ,"ONE-LINE" ,"STARS"]].head()
```

```python
def extract_info(s):
    s = s.replace('..', '').strip()
    info = {'Director': '', 'Stars': ''}

    parts = [p.strip() for p in s.split('|')]
    for i in parts:
        if i.startswith('Director:') or i.startswith('Directors:'):
            info['Director'] = i.split(':', 1)[1].strip()
        elif i.startswith('Star:') or i.startswith('Stars:'):
            info['Stars'] = i.split(':', 1)[1].strip()
    
    return pd.Series(info)

data[['Director', 'Stars']] = data['STARS'].apply(extract_info)

data[['STARS','Director','Stars']].head()
```

```python
data.drop(columns="STARS",inplace=True)
data.head()
```

```python
data['Director'] = data['Director'].replace('', pd.NA).fillna('Director Unknown')
data['Director'] = data['Director'].astype(str)
data.head()
```

```python
data['YEAR'] = data['YEAR'].astype(str)
data['YEAR'] = data['YEAR'].str.extract(r'((?:19|20)\d{2})')
data['YEAR'] = pd.to_datetime(data['YEAR'], format='%Y', errors='coerce').dt.year.astype('Int64')
data.head()
```

```python
data=data.sort_values(by='YEAR', ascending=False).reset_index(drop=True)
data
```

```python
print(data.duplicated().sum())
```

```python
data.drop_duplicates(inplace=True)
data.head()
```

```python
data['VOTES'] = data['VOTES'].astype(str).str.replace(',', '').str.strip()

data['VOTES'] = pd.to_numeric(data['VOTES'], errors='coerce')

null_count = data['VOTES'].isnull().sum()
print(null_count)
```

```python
data['VOTES']=data['VOTES'].fillna(0)
data
```

```python
data.drop(columns='Gross',inplace=True)
data
```

```python
data['GENRE'] = data['GENRE'].astype(str).str.replace('\n', '', regex=True).str.strip().str.replace(r'\s*,\s*', ', ', regex=True)
data
```

```python
data.to_csv('cleaned_data.csv', index=False)
```
